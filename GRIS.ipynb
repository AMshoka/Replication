```python
from numpy import random
import numpy as np
import itertools 
fL=1.4
f1=1.4
d=3
landaL=10**-6
lookup={}
```

```python
def findsubsets(s,n): 
    return list(itertools.combinations(s,n))
```

```python
def Create_Tasks(num,l):
    Set_length=num
    w, h = 3, Set_length;
    Matrix = [[0 for x in range(w)] for y in range(h)]
    x = random.uniform(0.1,100,size=(Set_length))
    x=np.sort(x)
    for i in range(Set_length):
        Matrix[i][0]=x[i]
        Matrix[i][1]=Set_length-i
        Matrix[i][2]=l #1.4GHz
        lookup["T"+str(i+1)]=tuple(Matrix[i])    
        
    return lookup
```

```python
def getList(dict): 
    list= [] 
    for key in dict.keys(): 
        list.append(key)
          
    return list
```

```python
def total_value(items, max_weight):
    return  sum([x[2] for x in items]) if sum([x[1] for x in items]) <= max_weight else 0
 
cache = {}
def solve(items, max_weight):
    if not items:
        return ()
    if (items,max_weight) not in cache:
        head = items[0]
        tail = items[1:]
        include = (head,) + solve(tail, max_weight - head[1])
        dont_include = solve(tail, max_weight)
        if total_value(include, max_weight) > total_value(dont_include, max_weight):
            answer = include
        else:
            answer = dont_include
        cache[(items,max_weight)] = answer
    return cache[(items,max_weight)]
```

```python
Succeful_Schedule_Based_on_Slack=0
Schedule_can_be_Fail=[]

for i in range(1000):
    
    print("#STEP"+str(i+1))
    lookup={}
    Tasks=Create_Tasks(10,fL)
    keys=getList(lookup) 
    items=[]
    for i in range(len(keys)):
        temp=[]
        temp.append(keys[i])
        temp1=lookup[keys[i]]
        temp.append(temp1[0])
        temp.append(temp1[1])
        items.append(tuple(temp))
    items=tuple(items)
    Slack=100
    while True:

        solution = solve(items, Slack)
        if(len(solution)==0):
            print("====>>Finished")
            print("====>>Remain Slack:",Slack)
            if (len(items)==0):
                Succeful_Schedule_Based_on_Slack=Succeful_Schedule_Based_on_Slack+1
                #print("************This Schedule executes Successful************")
            else:
                Schedule_can_be_Fail.append(items)
                print("====>>Task that can not recover"+str(items))
                print("************this Schedule can be failed************")

            break
        #print("***************")
        print("====>>Task Set:")
        print(items)
        print("====>>Remain Slack:",Slack)
        #print("***************")
        print("====>>The best tasks that can be retrieved and also comply with slack restrictions :")
        for x in solution:
            print (x[0])
        print ("Total priority:", total_value(solution, Slack))
        print ("Total Time for Task Recovery:", sum([x[1] for x in solution]))

        print('------------------------------------------------------------------------------------')
        subset=[]
        for j in range(len(solution)):
            subset.append(solution[j][0])
        items=list(items)
        remove=[]
        for i in range(len(subset)):
            for j in range(len(items)):
                if(subset[i]==items[j][0]):
                    fault_Prob = random.uniform(0,1)
                    if(fault_Prob<=10**-4):
                        Slack=Slack-items[j][1]
                        print(">>>>>>")
                        print("Oops, Task "+str(str(items[j])+"Failed"))
                        print("<<<<<<")
                    else:
                        remove.append(items[j])
        for i in range(len(remove)):
            items.remove(remove[i])
            
        items=tuple(items)

print("Probability of failure:",(1000-Succeful_Schedule_Based_on_Slack)/1000)
```
